```python
# Countries that has not given any rating to Restaurants Count
final_df[final_df['Rating text']=='Not rated'].groupby('Country').size().reset_index().rename(columns={0:'Non Rated Restaurants'})
```

Observation: <br>
India has the Highest No of Non Rated Restaurants

```python
# Currencies used by different Countries.
# Arranged in Descending Order
final_df.groupby(['Country','Currency']).size().reset_index().rename(columns={0:'Use Occurances'}).sort_values(by=['Use Occurances'], ascending=False).reset_index().drop('index',axis=1)
```

```python
# Countries categoried by Online Delivery
final_df.groupby(['Country','Has Online delivery']).size().reset_index().rename(columns={0: 'No of Restaurants'})
```

Observations: <br>
1. India and UAE has both No Delivery and Delivery Restaurants.
2. India has Most No of Delivery Restaurants and Non Delivery Restaurants.
3. India and UAE are the only Countries where Online Delivery is Available.

```python
# top 10 cuisines which are served at Restaurants
final_df[['Cuisines']].value_counts().reset_index().head(10).rename(columns={0:'No. of Restaurants'})
```

```python
cuisines_index= final_df[['Cuisines']].value_counts().index
cuisines_value= final_df[['Cuisines']].value_counts().values
```

Observations <br>

2. Chinese Food is Served after North Indian followed by Fast Food, Cafe and Bakery.
